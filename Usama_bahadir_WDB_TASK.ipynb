which will take the name of file and will return the 2 dataframes,

```python
# Function to read filename and return Countries as Column & Years as column
# Years are already in column so they can't be changed again


File_name = r'API_19_DS2_en_csv_v2_5346672.csv'
def fetching_data(File_name):
    
    
    """It will take the name of file and  extracts data of 3 countries Uk,sweden,switzerland.After preparing dataset,
    it will return 2 data frames .One will be original data frame  and second one will be its 
    transposed
    
    Parameters:
        filename: Will receive the name of file.
        
    Returns:
        will return 2 data frames.
        [DataFrame, Transposed of orginal data frame]: The original dataframe
        and its transposed version."""
    
        
    data_frame = pd.read_csv(File_name, skiprows = 4)
    data_frame = data_frame.set_index('Country Name')
    drop_columns = [
                'Country Code' , 
                'Indicator Name' , 
                'Indicator Code']
    data_frame = data_frame.drop(drop_columns, 
            axis = 1
            )
    data_frame_transpose = data_frame.T 
    return data_frame, data_frame_transpose

```

# . Fetching Data using wpdata API

```python
# I am fetching the dataset of the 3 countries of my choice using wbdata API

# I am interested in 3 countries 

# 1.SWEDEN(SWE)
# 2.SwitzerLand (ch)
# 3.United Kingdom (GBR)

interested_countries = ["SWE", "Ch", "GBR"]
 
# Indicators in which I am interested
required_indicators = {
                'EN.ATM.CO2E.SF.KT' : 'CO2 emissions from solid fuel consumption (kt)', 
                'AG.LND.AGRI.ZS' : 'Agricultural land (% of land area)', 
                'AG.LND.FRST.ZS' : 'Forest area (% of land area)', 
                'AG.LND.ARBL.ZS' : 'Arable land (% of land area)',
                'SP.POP.TOTL' : 'Total Population'
            }
```

```python
# Fetching the Required Dataset according to countries and indicators
data_frame = wbdata.get_dataframe(
                            required_indicators, 
                            country = interested_countries,
                            convert_date = False 
                        )

data = data_frame.unstack(level = 0)

data.head(2)
```

# Preparing the dataset of each country

We have multi index dataset , preparing the dataset separately for each country

```python
columns = []
sweden = pd.DataFrame()
switzerland = pd.DataFrame()
uk = pd.DataFrame()
for col in data.columns:
    if col[1] == "Sweden":
        dum = pd.DataFrame(data[col[0]]['Sweden'].values)
        sweden = pd.concat([sweden,dum], axis = 1)
    elif col[1]=="United Kingdom":
        dum = pd.DataFrame(data[col[0]]['United Kingdom'].values)
        uk = pd.concat([uk,dum],axis=1)
    else:
        dum = pd.DataFrame(data[col[0]]['Switzerland'].values)
        switzerland = pd.concat([switzerland, dum], axis = 1)
        columns.append(col[0])


switzerland.index = data.index
switzerland.columns = columns

sweden.index = data.index
sweden.columns = columns


uk.index = data.index
uk.columns = columns

```
